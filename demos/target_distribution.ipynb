```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook
```

```python
#Parameters to set
mu_x = -7.5
variance_x = 3

mu_y = -11
variance_y = 5

#Create grid and multivariate normal
x = np.linspace(-12.5,12.5,500)
y = np.linspace(-20,20,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]])

#Make a 3D plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos), cmap='viridis', linewidth=0, edgecolor='none')
ax.view_init(20, -135)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
zaxis_font = {'rotation':'30'}
ax.set_zlabel('Probability Density', **zaxis_font)
# label = ax.zaxis.label
# label.set_rotation(45)
ax.set_title('Probability of target locations')
plt.show()
plt.savefig("target-distribution.pdf")
```

```python
for i in range(10):
    x, y = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]]).rvs()
    print("Target", i+1, " position: ", str(round(x,3)) + ", " + str(round(y,3)))
```

```python
num_targets = 10
locs = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]]).rvs(size=num_targets)
for i in range(num_targets):
    print("Target", i+1, " position: ", str(round(locs[i][0],3)) + ", " + str(round(locs[i][1],3)))
```
